### function plot sum sales for each month


```python
import numpy as np
ax=np.arange(1,13)
ax
```

```python
def plotly(x):
    company=cars_data[cars_data['Company']==x]
    company['month']=company['Date'].dt.month
    sum_price_per_month=company.groupby('month')['price'].mean()
    yax=np.linspace(sum_price_per_month.min(),sum_price_per_month.max(),len(sum_price_per_month))
    plt.figure(figsize=(15,8))
    sum_price_per_month.plot(marker='o')
    plt.xticks(ax)
    plt.yticks(yax)
    plt.title(x)
    plt.ylabel('average of price per month')
    plt.show()
```

```python
for i in top_companies.keys():
    plotly(i)
```

```python
# function to calck percantege to every model from his sales

def calc(company):
    x=cars['Company']==company
    models=cars[x]['Model'].value_counts()
    length=len(cars[x])
    percantage=[(models[i]*100)/length for i in models.keys()]
#     result=''  
#     for i,j in enumerate(percantage):
#         result+=f"the model '{models.keys()[i]}' percantege of '{company}' company sales = {j} % \n"
    #return result
    plt.figure(figsize=(20,8))
    plt.bar(x=models.keys(),height=percantage)
    plt.title(f"Percentge of each model of sales '{company}' ")
    plt.ylabel('Percntge %')
    plt.xlabel('Company Models ')
    plt.xticks(models.keys(),rotation=45)
    plt.show()
```

```python
for i in top_companies.keys():
    calc(i)
```

### Clustering Data by price

```python
from sklearn.cluster import KMeans
```

```python
X=cars['price'].values.reshape(-1,1)

wcss=[]
for i in range(1,9):
    km=KMeans(i)
    km.fit(X)
    wcss_itr= km.inertia_
    wcss.append(wcss_itr)
```

```python
plt.plot(range(1,9),wcss,marker='o')
plt.xlabel('Number of cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(4)
km=kmeans.fit(X)
km.fit_predict(X)
cars['cls']=km.fit_predict(X)
```

```python
plt.figure(figsize=(20,5))
plt.scatter(cars['Model'],cars['price'],c=cars['cls'],cmap='rainbow',s=50)
plt.title('CLUSTER models of BY PRICE ')
plt.ylabel('PEICE')
plt.xlabel('Models')
plt.xticks([])
plt.legend()
```
